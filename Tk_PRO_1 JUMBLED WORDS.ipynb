# !!!!.......Jumbled Words Games.....!!!!

```python
import tkinter
from tkinter import*
```

```python
root = tkinter.Tk()
root.geometry('350x400+400+300')  #width x height
root.title('Jumbled ')  # title 
root.configure(background='#000000')  # for background color
#.......................................
import random 
from tkinter import messagebox
answer=['paris',
'mumbai',
'dubai',
'tokyo',
'colombo',
'delhi',
'tehran'
]


words=[
'pasir',
'baimum',
'baidu',
'tkyoo',
'cobolom',
'dleih',
'tnehra'
]

num =random.randrange(0,8,1)
#.....................................................


def res():
    global words,answer,num
    num = random.randrange(0,8,1)
    lbl.config(text=words[num])
    e1.delete(0,END)
#....................................
# Default function
def default():
    global words,answer,num
    lbl.config(text = words[num])
#....................................................
# check function
def checkans():
    global words,answer,num
    var=e1.get()
    if var == answer[num]:
        messagebox.showinfo('Success','This is a correct answer')
        res()
    else:
        messagebox.showerror('Error','This is a not correct answer')
        e1.delete(0,END)
    
#....................................................
####### LABEL

lbl= Label(
    root,
    text=' Your Here ',
    font =('cambria',18,'bold'),
    bg='#000000',
    fg='#ffffff',

)
lbl.pack(pady=30,ipady=10,ipadx=10)  #syntax to make it fix on location 
#.............................................................
 
####### ENTRY



ans1 =StringVar()
e1 = Entry (
    root,
    font =('cambria',16),
    textvariable = ans1,
)
e1.pack(ipady=5,ipadx=5)   # intrnal padding for y and x

#.........................................................
####### BUTTON


btncheck = Button (
    root,
    text = 'check',
    font = ('Comic sans ms',16),
    width = 16,
    bg='#8395A7',
    fg='#75DA8B',
    relief = GROOVE,
    command = checkans,
    
)
btncheck .pack(pady = 40)

#......................................

####### RESET BUTTON


btnreset = Button (
    root,
    text = 'Reset',
    font = ('Comic sans ms',16),
    width = 16,
    bg='#8395A7',
    fg='#FF4848',
    relief = GROOVE,
    command = res,
    
)
btnreset .pack()

#......................................

default()  # calling default funtion
root.mainloop()
```
